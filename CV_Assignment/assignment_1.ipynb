Q1. Given the image “lenna.png” with dimensions 220x220 pixels, apply bilinear and nearest
neighbor interpolation techniques to scale up the size of the image. After performing the 
interpolation, compare and contrast the visual results of the two methods in terms of 
smoothness, sharpness, and artifact appearance.

```python
# Importing all the necessary libraries
import matplotlib.pyplot as plt 
import cv2
import numpy as np
```

```python
img = cv2.imread("Lenna.png") # reading the image lenna.png
img_1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # converting the img form BGR to RGB so that i can plot it later
img.shape # printing the shape of the img
```

<h1>Nearest Neighbor Interpolation</h1>
<img src = "https://i0.wp.com/theailearner.com/wp-content/uploads/2018/10/Nearest_Neighbor.png?w=423&ssl=1" style= "display:grid;place-items: center;"/>

```python
resampled_img_NN = cv2.resize(img, (500,500), interpolation = cv2.INTER_NEAREST) # resizing image with nearest neighbor interpolation

cv2.imshow("Lenna",resampled_img_NN) # showing the image using opencv
cv2.waitKey(0)  
cv2.destroyAllWindows()

resampled_img_NN_RGB = cv2.cvtColor(resampled_img_NN, cv2.COLOR_BGR2RGB) # converting to RGB for plotting as matplotlib works on RGB and opencv works on BGR
plt.figure()
plt.title("Lenna image resampled with NN")
plt.imshow(resampled_img_NN_RGB)
plt.xticks([])
plt.yticks([])

plt.show()


```

```python
resampled_img_NN = resampled_img_NN.shape
resampled_img_NN # printing new resampled image shape
```

<h1>Bilinear interpolation</h1>
<img src = "https://i0.wp.com/theailearner.com/wp-content/uploads/2018/10/Bilinear_interpolation.png?w=423&ssl=1"/>
<img src="Screenshot 2024-09-20 170029.png"style = "height:40%;width:40%"/>

```python
resampled_img_Bilinear = cv2.resize(img, (500,500), interpolation = cv2.INTER_LINEAR) # rescaling using bilinear interpolation
resampled_img_Bilinear_RGB = cv2.cvtColor(resampled_img_Bilinear, cv2.COLOR_BGR2RGB)

cv2.imshow("Lenna_Bilinear",resampled_img_Bilinear)
cv2.imshow("Lenna NN",resampled_img_NN)
cv2.imshow("Lenna_original",img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.figure(figsize=(10,5))

plt.subplot(131)
plt.title("Lenna original")
plt.imshow(img_1)
plt.xlabel(img.shape)
plt.xticks([])
plt.yticks([])

plt.subplot(132)
plt.title("Lenna image resampled with NN")
plt.imshow(resampled_img_NN_RGB)
plt.xlabel(resampled_img_NN_RGB.shape)
plt.xticks([])
plt.yticks([])

plt.subplot(133)
plt.title("Lenna image resampled with Bilinear")
plt.imshow(resampled_img_Bilinear_RGB)
plt.xlabel(resampled_img_Bilinear_RGB.shape)
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()
```
