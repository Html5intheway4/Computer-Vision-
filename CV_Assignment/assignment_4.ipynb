## **Spatial domain averaging Filter**

<img src="Screenshot 2024-09-21 121950.png" style = "height:50%;width:40%"/>
<img src="Screenshot 2024-09-21 121930.png" style = "height:50%;width:40%"/>

```python
def apply_filter(image, kernel): # apply filter fuction takes image and kernel as an input
    # Getting dimensions of the image and the kernel
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape
    
    # Padding the image with zeros (padding needed to handle borders)
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant', constant_values=0)
    
    # Making an empty output image
    output_image = np.zeros_like(image)
    
    # Performing convolution
    for i in range(image_height):
        for j in range(image_width):
            # Extract the region of interest (ROI) from the padded image
            region = padded_image[i:i+kernel_height, j:j+kernel_width]
            
            # Element-wise multiplication of the region and the kernel, then sum the result
            output_value = np.sum(region * kernel)
            
            # Assign the result to the corresponding pixel in the output image
            output_image[i, j] = output_value
    
    return output_image

# Example usage
image = img_1
kernel = np.ones((5,5), np.float32) / 25

filtered_image = apply_filter(image, kernel)
filtered_image_car = apply_filter(img_2,kernel)


plt.figure()
plt.subplot(2,2,1)
plt.title("Original Noise.tif")
plt.imshow(img_1, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.title("Spatial domain averaging Filter")
plt.imshow(filtered_image, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.title("Original car.png")
plt.imshow(img_2, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
plt.title("spatial domain averaging Filter")
plt.imshow(filtered_image_car, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()

cv2.imshow('Filtered Image', filtered_image)
cv2.imshow('Filtered Image Car', filtered_image_car)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Gaussian Averaging Filter(Weighted averaging filter)**
<img src="Screenshot 2024-09-21 122413.png" style = "height:50%;width:40%"/>

```python
# Step 1: Create the Gaussian kernel
def gaussian_kernel(size, sigma=1):
    """Generates a Gaussian kernel."""
    kernel = np.fromfunction(
        lambda x, y: (1/(2*np.pi*sigma**2)) * np.exp(-((x-(size-1)/2)**2 + (y-(size-1)/2)**2) / (2*sigma**2)),
        (size, size)
    )
    return kernel / np.sum(kernel)  # Normalizing so that sum of all elements equals 1

def apply_filter(image, kernel):
    # Get dimensions of the image and the kernel
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape
    
    # Pad the image with zeros (padding needed to handle borders)
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant', constant_values=0)
    
    # Prepare an empty output image
    output_image = np.zeros_like(image)
    
    # Perform convolution
    for i in range(image_height):
        for j in range(image_width):
            # Extract the region of interest (ROI) from the padded image
            region = padded_image[i:i+kernel_height, j:j+kernel_width]
            
            # Element-wise multiplication of the region and the kernel, then sum the result
            output_value = np.sum(region * kernel)
            
            # Assign the result to the corresponding pixel in the output image
            output_image[i, j] = output_value
    
    return output_image

# Example usage
image = img_1
kernel_size = 19
sigma = 3.0

gaussian_k = gaussian_kernel(kernel_size, sigma)
filtered_image = apply_filter(image, gaussian_k)
filtered_image_car = apply_filter(img_2, gaussian_k)


plt.figure()
plt.subplot(2,2,1)
plt.title("Original Noise.tif")
plt.imshow(img_1, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.title("Gaussian averaging Filter")
plt.imshow(filtered_image, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.figure()
plt.subplot(2,2,3)
plt.title("Original Car.png")
plt.imshow(img_2, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
plt.title("Gaussian averaging Filter")
plt.imshow(filtered_image_car, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()

cv2.imshow('Filtered Image', filtered_image)
cv2.imshow('Filtered Image car', filtered_image_car)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Median averaging filter (3x3), (5,5)**

```python
def apply_median_filter(image, kernel_size):
    """Applies a median filter using a sliding window approach."""
    image_height, image_width = image.shape
    
    # Padding to handle border effects
    pad_size = kernel_size // 2
    padded_image = np.pad(image, pad_size, mode='constant', constant_values=0)
    
    # Preparing the output image
    output_image = np.zeros_like(image)
    
    # Perform median filtering
    for i in range(image_height):
        for j in range(image_width):
            # Extract the region of interest (ROI) around the current pixel
            region = padded_image[i:i+kernel_size, j:j+kernel_size]
            
            # Flatten the region, sort it, and find the median
            median_value = np.median(region)
            
            # Assign the median value to the central pixel in the output image
            output_image[i, j] = median_value
    
    return output_image

# Example usage
image = img_1  # Load image in grayscale
kernel_size_1 = 3  # Define the size of the kernel (should be odd)
kernel_size_2 = 5

# Apply the median filter
filtered_image_1 = apply_median_filter(image, kernel_size_1)
filtered_image_2 = apply_median_filter(image, kernel_size_2)
filtered_image_concat = np.hstack((filtered_image_1,filtered_image_2))
filtered_image_1_car = apply_median_filter(img_2, kernel_size_1)
filtered_image_2_car = apply_median_filter(img_2, kernel_size_2)
filtered_image_concat_car = np.hstack((filtered_image_1_car,filtered_image_2_car))

plt.figure(figsize=(10,5))
plt.subplot(2,3,1)
plt.title("Original Noise.tif")
plt.imshow(img_1, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,2)
plt.title("Median averaging Filter(3x3)")
plt.imshow(filtered_image_1, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,3)
plt.title("Median averaging Filter(5x5)")
plt.imshow(filtered_image_2, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,4)
plt.title("Original Car.png")
plt.imshow(img_2, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,5)
plt.title("Median averaging Filter(3x3)")
plt.imshow(filtered_image_1_car, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,3,6)
plt.title("Median averaging Filter(5x5)")
plt.imshow(filtered_image_2_car, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.tight_layout()
plt.show()

cv2.imshow('Filtered Image', filtered_image_concat)
cv2.imshow('Filtered Image car', filtered_image_concat_car)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
